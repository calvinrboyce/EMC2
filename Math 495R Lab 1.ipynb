```python
import numpy as np
from matplotlib import pyplot as plt
```

Problem 1

```python
x=np.linspace(-2*np.pi,2*np.pi,50)
y=np.sin(x)

plt.plot(x,y)
plt.xlim(-2*np.pi,2*np.pi)
plt.show()
```

```python
x=np.linspace(-2*np.pi,2*np.pi,50)
y=np.cos(x)

plt.plot(x,y)
plt.xlim(-2*np.pi,2*np.pi)
plt.show()
```

```python
x=np.linspace(-2*np.pi,2*np.pi,50)
y=np.arctan(x)

plt.plot(x,y)
plt.xlim(-2*np.pi,2*np.pi)
plt.show()
```

Problem 2

```python
x1=np.linspace(-2,.9999,50)
x2=np.linspace(1.0001,6,50)
y1=1/(x1-1)
y2=1/(x2-1)

plt.plot(x1,y1,'m--', linewidth=4)
plt.plot(x2,y2,'m--', linewidth=4)
plt.xlim(-2,6)
plt.ylim(-6,6)
plt.show()
```

Problem 3

```python
x1=np.linspace(0,100,100)
y1=x1

x2=np.linspace(0,100,100)
y2=10*np.sin(.5*x2)+x2

x3=np.linspace(0,100,100)
y3=100-(10*np.sin(.5*x3)+x3)

plt.plot(x1,y1,label='Python')
plt.plot(x2,y2,label='314')
plt.plot(x3,y3,label='341')

plt.ylabel('My Grades')
plt.xlabel('Time')
plt.legend(loc='upper left')
plt.title('This Semester')
plt.xlim(0,100)
plt.ylim(0,100)
plt.show()
```
