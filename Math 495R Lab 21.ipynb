Problem 2

```python
f=lambda x:x**2-1
def recursive_plot(x):
    x1=[i for i in range(50)]
    y1=[x]
    for i in range(1,50):
        y1.append(f(y1[i-1]))
    
    plt.plot(x1,y1,'.')
    plt.show
```
