```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Problem 1

```python
an=np.sin(np.linspace(1,1000,1000))

plt.plot(an,'.')
plt.show()
print("a) I'm guessing the limit points are everthing in [-1,1]")
print("b) No, we don't have any theorems about sin functions")
print("We don't know if there is a value of sin(n)<e")

def subsequence_close_to_limit(A,L,E):
    subsequence=[]
    for i in range(len(A)):
        if abs(A[i]-L)<E:
            subsequence.append(i)
    subsequence_values=[A[i] for i in subsequence]
    
    plt.plot(A,'.')
    plt.plot(subsequence,subsequence_values,'.')
    plt.show
    return subsequence

print("c)")
bn=np.sin(np.linspace(1,100000,100000))
subsequence_close_to_limit(bn,.25,.0001);

print("d) I still think they're all limit points")
```

```python
def enumerate_rationals(n):
    rationals=[0]
    for i in range(1,n+1):
        for j in range(1,i+1):
            if math.gcd(i,j)==1:
                rationals.append(j/i)
    return rationals

r=enumerate_rationals(100)
subsequence_close_to_limit(r,.2734909,.001)
print("We can prove that every point (rational or irrational) in [0,1] is a limit point, because for every e>0, there exists a rational number between x and x+e")
```
