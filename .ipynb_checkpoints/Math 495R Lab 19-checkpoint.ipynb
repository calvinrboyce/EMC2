```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

Problem 1

```python
x = np.arange(-10,10,1)
y = np.arange(-10,10,1)

X,Y = np.meshgrid(x,y)
u=Y**2-2*X*Y
v=3*X*Y-6*X**2

fig,ax=plt.subplots(figsize=(5,5))

ax.quiver(X,Y,u,v)
ax.set_aspect('equal')
plt.show()
```

Problem 2

```python
x = np.arange(-1,1,.1)
y = np.arange(-1,1,.1)

X,Y = np.meshgrid(x,y)
r=(X**2+Y**2)**.5
u=(r**2-2*r)*X
v=(r**2-2*r)*Y

fig,ax=plt.subplots(figsize=(5,5))
ax.quiver(X,Y,u,v)
ax.set_aspect('equal')
plt.show()

x = np.arange(-5,5,.5)
y = np.arange(-5,5,.5)

X,Y = np.meshgrid(x,y)
r=(X**2+Y**2)**.5
u=(r**2-2*r)*X
v=(r**2-2*r)*Y

fig,ax=plt.subplots(figsize=(5,5))
ax.quiver(X,Y,u,v)
ax.set_aspect('equal')
plt.show()
```

Problem 3a

```python
x = np.arange(-10,10,1)
y = np.arange(-10,10,1)

X, Y = np.meshgrid(x, y)
Z = np.log(1+X**2+2*Y**2)
dy, dx = np.gradient(Z)

fig, ax = plt.subplots(figsize=(5,5))
ax.quiver(X,Y,dx,dy)
ax.set_aspect('equal')
plt.contour(X,Y,Z,10,cmap='coolwarm')

plt.show()
```

Problem 3b

```python
x = np.arange(-5,5,.5)
y = np.arange(-5,5,.5)

X, Y = np.meshgrid(x, y)
Z = np.cos(X)-2*np.sin(Y)
dy, dx = np.gradient(Z)

fig, ax = plt.subplots(figsize=(5,5))
ax.quiver(X,Y,dx,dy)
ax.set_aspect('equal')
plt.contour(X,Y,Z,10,cmap='coolwarm')

plt.show()
```

Problem 4a

```python
ax = plt.figure().add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-2, 2, 1),
                      np.arange(-2, 2, 1),
                      np.arange(-2, 2, 1))

# Make the direction data for the arrows
u = 1
v = 2
w = 3

ax.quiver(x, y, z, u, v, w, length=0.5, normalize=True)

plt.show()
```

Problem 4d

```python
ax = plt.figure().add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-4, 4, 1),
                      np.arange(-4, 4, 1),
                      np.arange(-2, 2, 1))

# Make the direction data for the arrows
u = np.sin(y)
v = x*np.cos(y)+np.cos(z)
w = -y*np.sin(z)

ax.quiver(x, y, z, u, v, w, length=0.5, normalize=True)

plt.show()
```
