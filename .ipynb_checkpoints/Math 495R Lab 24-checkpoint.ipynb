Problem 2

```python
x=sy.symbols('x')
expr=x**3-1
f=sy.lambdify(x,expr)
df=sy.lambdify(x,sy.diff(expr))
```

Problem 4

```python
def basins_of_attraction(f,Df,zeros,ranges,res,iters):
    x_real=np.linspace(ranges[0],ranges[1],res)
    x_imag=np.linspace(ranges[2],ranges[3],res)
    X_real,X_imag=np.meshgrid(x_real,x_imag)
    X_0=X_real+1j*X_imag
    
    for i in range(iters):
        X_0=-f(X_0)/Df(X_0)+X_0
        
    Y=[[np.argmin(abs(np.array(zeros)-c)) for c in d] for d in X_0]
    
    plt.pcolormesh(X_real,X_imag,Y,cmap='brg')
    plt.show()
```

Problem 5

```python
f1=lambda x:x**3-1
df1=lambda x:3*x**2
basins_of_attraction(f1,df1,[1,-1/2+np.sqrt(3)/2*1j,-1/2-np.sqrt(3)/2*1j],[-1.5,1.5,-1.5,1.5],500,100)
```

```python
g=lambda x:x**3-x
dg=lambda x:3*x**2-1
basins_of_attraction(g,dg,[0,1,-1],[-1.5,1.5,-1.5,1.5],500,100)
```

Problem 6

```python
h=lambda x:x**4-x
dh=lambda x:4*x**3-1
basins_of_attraction(h,dh,[0,1,-1/2-np.sqrt(3)/2*1j,-1/2+np.sqrt(3)/2*1j],[-1.5,1.5,-1.5,1.5],500,100)
```
