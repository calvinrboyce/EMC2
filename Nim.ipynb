```python
import random
```

```python
#check to see if a move is valid
def checkmove(position,pile,remove):
    if position[pile-1]-remove<0:
        return False
    else:
        return True
```

```python
#one player's move
def player(n,position):
    valid=False
    while not valid:
        move=input('Player '+str(n)+', what move would you like to make?')
        pile,remove=move.split()
        pile,remove=int(pile),int(remove)
        if checkmove(position,pile,remove):
            valid=True
        else:
            print('That input was invalid. Please enter a valid move.')
    
    position[pile-1]-=remove
```

```python
#play the game
def playgame(strategy):
    while True:
        #should we play a game?
        valid=False
        while not valid:
            game_start=input('Would you like to play a game? (Y/N)')
            if game_start=='y' or game_start=='Y':
                valid=True
            elif game_start=='n' or game_start=='N':
                return 'game over'
            else:
                print('That input was invalid.')

        #set players
        valid=False
        P1=''
        P2=''
        while not valid:
            P1=input('Player 1: Computer (C) or Human (H)? (C/H)')
            if P1=='C' or P1=='H':
                valid=True
            else:
                print('That input was invalid.')
        valid=False
        while not valid:
            P2=input('Player 2: Computer (C) or Human (H)? (C/H)')
            if P2=='C' or P2=='H':
                valid=True
            else:
                print('That input was invalid.')
        players=[P1,P2]
        moves=[[],[]]

        #for PvP or PvC
        if P1=='H' or P2=='H':
            position=[7,5,3]
            n=1
            while True:
                print('The piles have '+str(position[0])+', '+str(position[1])+', '+str(position[2])+' objects')
                if players[n-1]=='H':
                    player(n,position)
                else:
                    computer(n,position,moves,strategy)
                
                if position==[0,0,0]:
                    print('Player '+str(n)+' wins!')
                    break
                n=2-((n+1)%2)
                
        #for CvC
        if P1=='C' and P2=='C':
            game_num=int(input('How many games?'))
            for i in range(1,game_num+1):
                position=[7,5,3]
                n=1
                print('--Beginning Game '+str(i)+'--')
                while True:
                    print('The piles have '+str(position[0])+', '+str(position[1])+', '+str(position[2])+' objects')
                    computer(n,position,moves,strategy)

                    if position==[0,0,0]:
                        print('Player '+str(n)+' wins!')
                        break
                    n=2-((n+1)%2)
            print('--Completed '+str(game_num)+' games--')
```

```python
def initialize():
    strategy=[]
    for i in range(754):
        a=i//100
        b=i//10%10
        c=i%10
        digits=[a,b,c]
        
        if b>5 or c>3:
            strategy.append([])
            continue
        
        elem=[]
        for j in range(1,4):
            for k in range(digits[j-1],0,-1):
                elem.append([j,k,50])
        strategy.append(elem)
    return strategy

strategy=initialize()
strategy1=initialize()
```

```python
def computer(n,position,moves,strategy):
    pos=position[0]*100+position[1]*10+position[2]
    
    #find the max strategy value
    mx=strategy[pos][0][2]
    for i in strategy[pos]:
        if i[2]>mx:
            mx=i[2]
    
    #find the moves with max strategy
    bestmoves=[]
    for i in range(len(strategy[pos])):
        if strategy[pos][i][2]==mx:
            bestmoves.append(i)
    
    #pick a move at random
    m=bestmoves[random.randint(0,len(bestmoves)-1)]
    
    #execute that move
    a,b,trash=strategy[pos][m]
    position[a-1]-=b
    print('Computer player '+str(n)+' takes '+str(b)+' objects from pile '+str(a))
    
    #store the move
    moves[n-1].append([pos,m])
```

```python
def computertrainer(n,position,moves,strategy):
    while True:
        pos=position[0]*100+position[1]*10+position[2]

        #find the max strategy value
        mx=strategy[pos][0][2]
        for i in strategy[pos]:
            if i[2]>mx:
                mx=i[2]

        #find the moves with max strategy
        bestmoves=[]
        for i in range(len(strategy[pos])):
            if strategy[pos][i][2]>mx-90:
                bestmoves.append(i)

        #pick a move at random
        m=bestmoves[random.randint(0,len(bestmoves)-1)]

        #execute that move
        a,b,trash=strategy[pos][m]
        position[a-1]-=b

        #store the move
        moves[n-1].append([pos,m])

        n=2-((n+1)%2)
        if position==[0,0,0]:
            return 1-((n+1)%2),n-1
```

```python
def traincomputer(strategy):
    for i in range(15000):
        position=[7,5,3]
        moves=[[],[]]
        w,l=computertrainer(1,position,moves,strategy)

        #adjust winning moves
        for i in moves[w]:
            strategy[i[0]][i[1]][2]+=29
            if strategy[i[0]][i[1]][2]>100:
                strategy[i[0]][i[1]][2]==100

        #adjust losing moves
        for i in moves[l]:
            strategy[i[0]][i[1]][2]-=11
            if strategy[i[0]][i[1]][2]<0:
                strategy[i[0]][i[1]][2]==0

        #adjust final moves
        wm=moves[w][len(moves[w])-1]
        strategy[wm[0]][wm[1]][2]+=1000
        lm=moves[l][len(moves[l])-1]
        strategy[lm[0]][lm[1]][2]-=1000
        
traincomputer(strategy1)
```
