```python
import matplotlib.pyplot as plt
import math as math
import numpy as np

N=20
a=[]
for n in range(0,N):
    a.append(((-1.0)**n)*n/(2.0*n+1.0))

plt.plot(a,".")
plt.show()
```

```python
import functools
@functools.lru_cache(maxsize=256)
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)

x=[n for n in range(2,100)]
y=[fib(n)/fib(n-1) for n in range(2,100)]

y2=[1.618 for n in range(2,100)]

plt.plot(x,y,'.')
plt.plot(x,y2)
plt.xlim(0,100)
plt.ylim(1.616,1.620)
plt.show()
print('I conjecture the limit is 1.618')
```
