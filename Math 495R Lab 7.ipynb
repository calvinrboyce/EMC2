```python
import matplotlib.pyplot as plt
import math
```

```python
curr_sum=0
partial_sums=[]

for n in range(0,50):
    curr_sum +=1/2**n
    partial_sums.append(curr_sum)
    
plt.plot(partial_sums,'.')
plt.show()
#converging to the right limit!
```

```python
curr_sum=0
partial_sums=[]

for n in range(0,50):
    curr_sum +=(-1)**n
    partial_sums.append(curr_sum)
    
plt.plot(partial_sums,'.')
plt.show()
#does not converge
```

```python
curr_sum=0
partial_sums=[]

for n in range(1,1000):
    curr_sum +=1/n
    partial_sums.append(curr_sum)
    
plt.plot(partial_sums,'.')
plt.show()
#diverging
```

```python
curr_sum=0
partial_sums=[]

for n in range(1,10000):
    curr_sum +=1/(n*math.log(n+1,10))
    partial_sums.append(curr_sum)
    
plt.plot(partial_sums,'.')
plt.show()
#looks like this one diverges as well
```

```python
curr_sum=-1/math.sqrt(3)
partial_sums=[]

for n in range(2,1000000):
    curr_sum +=((-1)**n)*(math.floor(math.log(n,2))/n)
    partial_sums.append(curr_sum)

plt.plot(partial_sums,'.')
plt.ylim(-0.00025,0)
plt.xlim(800000,1000000)
plt.show()
# This one seems to converge to -0.00015
```
